```python
import pandas as pd

fandango = pd.read_csv('fandango_score_comparison.csv')
fandango.columns
fandango.head()
```

```python
fandango.sort_values('FILM',inplace=True)
fandango.head()
```

```python
fandango.columns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.scatter(fandango['RT_norm'],fandango['IMDB_norm'])
ax1.set_xlabel('RT_norm')
ax1.set_ylabel('IMDB_norm')
ax1.set_xlim(0,5)
fig.subplots_adjust(right=1.1,top=1)
ax2.scatter(fandango['RT_norm'],fandango['Fandango_Ratingvalue'])
ax2.set_xlabel('RT_norm')
ax2.set_ylabel('Fandango_Ratingvalue')
ax2.set_xlim(0,5)

plt.show()
```

From the above visual it is clear that 

1.There exists a positive linear correlation between IMDB_norm and RT_norm.

2.There is no correlation between Fandango_rating_value and RT_norm

```python
fig = plt.figure()
ax3 = fig.add_subplot(121)
ax4 = fig.add_subplot(122)
ax3.scatter(fandango['RT_norm'],fandango['Metacritic_norm_round'])
ax3.set_xlabel('RT_norm')
ax3.set_ylabel('Metacritic_norm_round')
ax3.set_xlim(0,5)

ax4.scatter(fandango['Fandango_Ratingvalue'],fandango['Metacritic_norm_round'])
ax4.set_xlabel('Fandango_Ratingvalue')
ax4.set_ylabel('Metacritic_norm_round')
ax4.set_xlim(0,5)
```

From the above visual it is clear that

1. There exists a positive correlation between RT_norm and Metacritic_norm_round

2. There exists no correlation between Metacritic_norm_round and Fandango_Ratingvalue
